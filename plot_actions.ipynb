```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
P1 = pd.read_csv("recordings/bs521.csv", header=0, index_col=0)
P2 = pd.read_csv("recordings/bs522.csv", header=0, index_col=0)
print(P1)

dfs = [P1, P2]
labels=["Seed 521", "Seed 522", ]
```

```python
# def plotActions(dfs:list, labels:list):
fig, ax = plt.subplots(3, 2, figsize=(9,5), sharex=True, sharey=True)
for df, label in zip(dfs, labels):
    df = df.iloc[250:350]
    ax[0][0].plot(df.index, df.bthigh, label=label)
    ax[0][0].set_title("Back Thigh")
    ax[0][1].plot(df.index, df.fthigh, label=label)
    ax[0][1].set_title("Front Thigh")

    ax[1][0].plot(df.index, df.bshin, label=label)
    ax[1][0].set_title("Back Shin")
    ax[1][1].plot(df.index, df.fshin, label=label)
    ax[1][1].set_title("Front Shin")

    ax[2][0].plot(df.index, df.bfoot, label=label)
    ax[2][0].set_title("Back Foot")
    ax[2][1].plot(df.index, df.bfoot, label=label)
    ax[2][1].set_title("Front Foot")

    ax[2][0].set_xlabel("step")
    ax[2][1].set_xlabel("step")

    ax[0][0].set_ylabel("Angle [rad]")
    ax[1][0].set_ylabel("Angle [rad]")
    ax[2][0].set_ylabel("Angle [rad]")

    ax[0][0].legend(ncol=2, loc="lower right", fontsize=8)

    for ax_ in fig.get_axes():
        ax_.grid(True)
        ax_.set_ylim(-1.3, 1.3)


fig.set_tight_layout(True)
plt.show()
```
