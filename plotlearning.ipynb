```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### basic model

```python
file1 = "PPOv1/basicSeed520.csv"
file2 = "PPOv1/basicSeed521.csv"
file3 = "PPOv1/basicSeed522.csv"

sample_every = 1
P1 = pd.read_csv(file1, index_col=0, header=0).iloc[::sample_every]
P2 = pd.read_csv(file2, index_col=0, header=0).iloc[::sample_every]
P3 = pd.read_csv(file3, index_col=0, header=0).iloc[::sample_every]
# print(len(P1))
# P = pd.merge(P1, P2, on="timestep", how="outer").sort_values("timestep")
# P = pd.merge(P, P3, on="timestep", how="outer").sort_values("timestep")
# P = P.fillna(method='ffill')
# print(len(P))
# print(P)

P = pd.concat([P1, P2, P3]).sort_values("timestep")
```

```python
fig, ax = plt.subplots(3, 1, figsize=(9, 7), sharex=True)

ax[0].plot(P1.timestep, P1["reward"], linewidth=2.5, color="tab:red")#, linestyle="-")
ax[1].plot(P2.timestep, P2["reward"], linewidth=2.5, color="tab:green")#, linestyle="-.")
ax[2].plot(P3.timestep, P3["reward"], linewidth=2.5, color="tab:blue")#, linestyle="--")
ax[0].set_title(f"Seed 520 | Final Reward: {round(P1.reward.iloc[-1],2)}")
ax[1].set_title(f"Seed 521 | Final Reward: {round(P2.reward.iloc[-1],2)}")
ax[2].set_title(f"Seed 522 | Final Reward: {round(P3.reward.iloc[-1],2)}")
for ax_ in fig.get_axes():
    ax_.set_ylim(-200, 2000)
    ax_.set_ylabel("Reward")
    ax_.grid()

ax[2].set_xlabel("Timestep")

fig.set_tight_layout(True)
```

```python
sns.lineplot(data=P, x="timestep", y="reward")
```
