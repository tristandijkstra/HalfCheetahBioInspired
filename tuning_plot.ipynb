```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base1 = pd.read_csv("SB3/baseline_PPO_41.csv", header=0, index_col=0)
base2 = pd.read_csv("SB3/baseline_PPO_42.csv", header=0, index_col=0)
highlr1 = pd.read_csv("SB3/highlr_PPO_1.csv", header=0, index_col=0)
highlr2 = pd.read_csv("SB3/highlr_PPO_2.csv", header=0, index_col=0)
lowlr1 = pd.read_csv("SB3/lowlr_PPO_1.csv", header=0, index_col=0)
lowlr2 = pd.read_csv("SB3/lowlr_PPO_3.csv", header=0, index_col=0)

highclip1 = pd.read_csv("SB3/highclip_PPO_1.csv", header=0, index_col=0)
highclip2 = pd.read_csv("SB3/highclip_PPO_2.csv", header=0, index_col=0)

highGAE1 = pd.read_csv("SB3/highGAE_PPO_1.csv", header=0, index_col=0)
highGAE2 = pd.read_csv("SB3/highGAE_PPO_2.csv", header=0, index_col=0)
lowGAE1 = pd.read_csv("SB3/lowGAE_PPO_1.csv", header=0, index_col=0)
lowGAE2 = pd.read_csv("SB3/lowGAE_PPO_2.csv", header=0, index_col=0)

gamma1 = pd.read_csv("SB3/gamma_PPO_1.csv", header=0, index_col=0)
gamma2 = pd.read_csv("SB3/gamma_PPO_2.csv", header=0, index_col=0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9,3))

ax.plot(base1.Step, base1.Value, color="tab:blue", label=r"baseline ($lr=2.5E-4$)")
ax.plot(base2.Step, base2.Value, color="tab:blue")
ax.fill_between(base1.Step, base1.Value, base2.Value, color="tab:blue", alpha=0.2)

ax.plot(highlr1.Step, highlr1.Value, color="tab:orange", label=r"high learning rate ($lr=2.5E-3$)", linestyle='--')
ax.plot(highlr2.Step, highlr2.Value, color="tab:orange", linestyle='--')
ax.fill_between(highlr1.Step, highlr1.Value, highlr2.Value, color="tab:orange", alpha=0.2)

ax.plot(lowlr1.Step, lowlr1.Value, color="tab:green", label=r"low learning rate ($lr=2.5E-5$)", linestyle='-.')
ax.plot(lowlr2.Step, lowlr2.Value, color="tab:green", linestyle='-.')
ax.fill_between(lowlr1.Step, lowlr1.Value, lowlr2.Value, color="tab:green", alpha=0.2)

ax.set_xlim(0, 4e5+500)
ax.set_ylim(-500, 2500)

ax.set_ylabel("Reward")
ax.set_xlabel("Timestep")
ax.grid()
ax.legend()
fig.set_tight_layout(True)

fig.savefig("plots/tunelr.png")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9,3))

ax.plot(lowlr1.Step, lowlr1.Value, color="tab:blue", label=r"New Baseline (CR=0.1)")
ax.plot(lowlr2.Step, lowlr2.Value, color="tab:blue")
ax.fill_between(lowlr1.Step, lowlr1.Value, lowlr2.Value, color="tab:blue", alpha=0.2)

ax.plot(highclip1.Step, highclip1.Value, color="tab:orange", label=r"High Clip Range (CR=0.2)", linestyle='--')
ax.plot(highclip2.Step, highclip2.Value, color="tab:orange", linestyle='--')
ax.fill_between(highclip1.Step, highclip1.Value, highclip2.Value, color="tab:orange", alpha=0.2)


ax.set_xlim(0, 4e5+500)
ax.set_ylim(-500, 2500)

ax.set_ylabel("Reward")
ax.set_xlabel("Timestep")
ax.grid()
ax.legend()
fig.set_tight_layout(True)

fig.savefig("plots/tuneCR.png")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9,3))

ax.plot(highclip1.Step, highclip1.Value, color="tab:blue", label=r"New Baseline (lambda=0.92)")
ax.plot(highclip2.Step, highclip2.Value, color="tab:blue")
ax.fill_between(highclip1.Step, highclip1.Value, highclip2.Value, color="tab:blue", alpha=0.2)

ax.plot(highGAE1.Step, highGAE1.Value, color="tab:orange", label=r"High GAE (lambda=0.96)", linestyle='-.')
ax.plot(highGAE2.Step, highGAE2.Value, color="tab:orange", linestyle='-.')
ax.fill_between(highGAE1.Step, highGAE1.Value, highGAE2.Value, color="tab:orange", alpha=0.2)

ax.plot(lowGAE1.Step, lowGAE1.Value, color="tab:green", label=r"Low GAE (lambda=0.88)", linestyle='--')
ax.plot(lowGAE2.Step, lowGAE2.Value, color="tab:green", linestyle='--')
ax.fill_between(lowGAE1.Step, lowGAE1.Value, lowGAE2.Value, color="tab:green", alpha=0.2)


ax.set_xlim(0, 4e5+500)
ax.set_ylim(-500, 2500)

ax.set_ylabel("Reward")
ax.set_xlabel("Timestep")
ax.grid()
ax.legend()
fig.set_tight_layout(True)

fig.savefig("plots/tuneGAE.png")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9,3))

ax.plot(highclip1.Step, highclip1.Value, color="tab:blue", label=r"New Baseline (gamma=0.99)")
ax.plot(highclip2.Step, highclip2.Value, color="tab:blue")
ax.fill_between(highclip1.Step, highclip1.Value, highclip2.Value, color="tab:blue", alpha=0.2)

ax.plot(gamma1.Step, gamma1.Value, color="tab:orange", label=r"Lower Discount Factor (gamma = 0.98)", linestyle='-.')
ax.plot(gamma2.Step, gamma2.Value, color="tab:orange", linestyle='-.')
ax.fill_between(gamma1.Step, gamma1.Value, gamma2.Value, color="tab:orange", alpha=0.2)


ax.set_xlim(0, 4e5+500)
ax.set_ylim(-500, 2500)

ax.set_ylabel("Reward")
ax.set_xlabel("Timestep")
ax.grid()
ax.legend()
fig.set_tight_layout(True)

fig.savefig("plots/tuneGamma.png")
```
